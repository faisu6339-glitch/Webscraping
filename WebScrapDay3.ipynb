```python
from numpy._core.defchararray import find
data=[]
for link in df['Book Link']:
  res=requests.get(link)
  soup=BeautifulSoup(res.text,'html.parser')
  typ=soup.find('ul' , class_='breadcrumb').find_all('a')[2].text

  upc=soup.find('table',class_='table table-striped').find_all('td')[0].text[2:]
  price_x= soup.find('table',class_='table table-striped').find_all('td')[2].text[2:]
  price_1=soup.find('table',class_='table table-striped').find_all('td')[3].text[2:]
  tax=soup.find('table',class_='table table-striped').find_all('td')[4].text
  avail=soup.find('table',class_='table table-striped').find_all('td')[5].text
  qn=soup.find('table',class_='table table-striped').find_all('td')[6].text

  data.append([typ,upc,price_x,price_1,tax,avail,qn])
```

```python
df=pd.DataFrame(data,columns=['Type','UPC','Price (Excl. Tax)','Price (Incl. Tax)','Tax','Availability','Number of Reviews'])
```
