```python
soup=BeautifulSoup(res.text,'html.parser')
```

```python
soup
```

```python
quotes_list = []

for q in soup.find_all('div', class_='quote'):
    text = q.find('span', class_='text').text[1:-1]
    quotes_list.append(text)

quotes = "\n".join(quotes_list)   # line-by-line string

print(quotes)
```

```python
all_quotes = []

for block in soup.find_all('div', class_='quote'):

    # Quote text
    quote_text = block.find('span', class_='text').text[1:-1]

    # Author
    author = block.find('small', class_='author').text

    # About link
    about_link = block.find('a')['href']

    # Tags (the <a class="tag"> elements)
    tags = [tag.text for tag in block.find_all('a', class_='tag')]

    # Keywords (inside <meta class="keywords">)
    keywords = block.find('meta', class_='keywords')['content'].split(',')

    # Store everything in a dictionary
    all_quotes.append({
        "quote": quote_text,
        "author": author,
        "about": about_link,
        "tags": tags,
        "keywords": keywords
    })

# Show sample
all_quotes[:1]   # show first 1 items
```

```python
from tqdm import tqdm
```

```python
data=[]
for page in tqdm(range(1,11)):
    link=f"https://quotes.toscrape.com/page/{page}/"
    res=requests.get(link)
    soup=BeautifulSoup(res.text,'html.parser')

    for sp in soup.find_all('div',class_='quote'):
        quote=sp.find('span',class_='text').text[1:-1]
        author=sp.find('small',class_='author').text
        about_link=sp.find('a')['href']
        author_id=sp.find('a')['href'].split('/')[2].split('-')
        tags=[tag.text for tag in sp.find_all('a',class_='tag')]
        keywords=sp.find('meta',class_='keywords')['content'].split(',')

        data.append([quote,author,about_link,author_id,tags])
```
