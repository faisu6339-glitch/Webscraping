#Accessing the thumbnail of the videos


```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(driver.page_source, 'html.parser')

sp = soup.find_all('ytd-grid-video-renderer')

print(type(sp[0]))   # MUST show: <class 'bs4.element.Tag'>
```

```python
thumbnails = []

for video in sp:
    title_tag = video.find('a', id='video-title')

    if title_tag and title_tag.get('href'):
        video_id = title_tag['href'].split('v=')[-1]
        thumb = f"https://img.youtube.com/vi/{video_id}/maxresdefault.jpg"
    else:
        thumb = None

    thumbnails.append(thumb)

print(thumbnails[:5])
```

```python
print(type(sp))
print(type(sp[0]))
```

```python
data = []

for video in sp:
    # ---------- TITLE + LINK ----------
    title_tag = video.find('a', id='video-title')

    if title_tag and title_tag.get('href'):
        title = title_tag.text.strip()
        video_link = "https://www.youtube.com" + title_tag['href']

        # video id for thumbnail
        video_id = title_tag['href'].split('v=')[-1]
        thumbnail = f"https://img.youtube.com/vi/{video_id}/maxresdefault.jpg"
    else:
        title = None
        video_link = None
        thumbnail = None

    # ---------- VIEWS + DATE ----------
    meta = video.find_all('span', class_='style-scope ytd-grid-video-renderer')

    views = meta[0].text.strip() if len(meta) > 0 else None
    date_time = meta[1].text.strip() if len(meta) > 1 else None

    # ---------- STORE DATA ----------
    data.append([
        title,
        video_link,
        views,
        date_time,
        thumbnail
    ])
```

#âœ… STEP 1: Auto-scroll the Videos page

```python
import time

last_height = driver.execute_script("return document.documentElement.scrollHeight")

while True:
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(2)

    new_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height

print("âœ… Scrolling completed")
```

#âœ… STEP 2: Re-parse page after scrolling

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(driver.page_source, 'html.parser')
sp = soup.find_all('ytd-grid-video-renderer')

print("Total videos found:", len(sp))
```

#âœ… STEP 3: Full scrape (title, link, views, date, thumbnail)

```python
data = []

for video in sp:
    # title + link + thumbnail
    title_tag = video.find('a', id='video-title')

    if title_tag and title_tag.get('href'):
        title = title_tag.text.strip()
        video_link = "https://www.youtube.com" + title_tag['href']
        video_id = title_tag['href'].split('v=')[-1]
        thumbnail = f"https://img.youtube.com/vi/{video_id}/maxresdefault.jpg"
    else:
        title = None
        video_link = None
        thumbnail = None

    # views + upload time
    meta = video.find_all('span', class_='style-scope ytd-grid-video-renderer')
    views = meta[0].text.strip() if len(meta) > 0 else None
    date_time = meta[1].text.strip() if len(meta) > 1 else None

    data.append([title, video_link, views, date_time, thumbnail])
```

```python
import pandas as pd

df = pd.DataFrame(
    data,
    columns=["Title", "Video_Link", "Views", "Upload_Time", "Thumbnail"]
)

df.head()
```
