```python
 quotes_list = []
authors_list = []

for q in soup.find_all('div', class_='quote'):
    text = q.find('span', class_='text').text[1:-1]     # the quote
    author = q.find('small', class_='author').text      # the author

    quotes_list.append(text)
    authors_list.append(author)

# Show result
for quote, author in zip(quotes_list, authors_list):
    print(quote)
    print("—", author)
    print()
```

```python
all_quotes = []

for block in soup.find_all('div', class_='quote'):

    # Quote text
    quote_text = block.find('span', class_='text').text[1:-1]

    # Author
    author = block.find('small', class_='author').text

    # About link
    about_link = block.find('a')['href']

    # Tags (the <a class="tag"> elements)
    tags = [tag.text for tag in block.find_all('a', class_='tag')]

    # Keywords (inside <meta class="keywords">)
    keywords = block.find('meta', class_='keywords')['content'].split(',')

    # Store everything in a dictionary
    all_quotes.append({
        "quote": quote_text,
        "author": author,
        "about": about_link,
        "tags": tags,
        "keywords": keywords
    })

# Show sample
all_quotes[:1]   # show first 1 items
```

```python
for item in all_quotes:
    print("Quote:", item["quote"])
    print("Author:", item["author"])
    print("About:", item["about"])
    print("Tags:", ", ".join(item["tags"]))
    print("Keywords:", ", ".join(item["keywords"]))
    print("-" * 60)
```

Accessing Single Data

```python
data=[]
for sp in soup.find_all('div', class_='quote'):
    quote=sp.find('span',class_='text').text[1:-1]
    author=sp.find('small',class_='author').text
    author_id=sp.find('a')['href']
    about=sp.find('a')['href']
    tags=[]
    for tag in sp.find_all('a',class_='tag'):
        tags.append(tag.text)
    tags=','.join(tags)
    data.append([quote,author,author_id,about,tags])
```

```python
df=pd.DataFrame(data,columns=['quote','author','author_id','about','tags'])
```
